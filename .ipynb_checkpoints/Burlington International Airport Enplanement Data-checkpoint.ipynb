```python
#Move the year into the index instead of as its own column
wholedf = wholedf.set_index('Year')

#Delete the totals column for easier plotting
del wholedf['Totals ']
```

Departure Distribution
-----------------------
The dataset has departure information by month since 1988 by month, but what does the distribution look like across the month for that entire time?  Do some months categorically have more, like around the holidays?

```python
sns.kdeplot(wholedf['January ']);
sns.kdeplot(wholedf['February']);
sns.kdeplot(wholedf['March']);
sns.kdeplot(wholedf['April']);
sns.kdeplot(wholedf['May']);
sns.kdeplot(wholedf['June']);
sns.kdeplot(wholedf['July']);
sns.kdeplot(wholedf['August']);
sns.kdeplot(wholedf['September']);
sns.kdeplot(wholedf['October']);
sns.kdeplot(wholedf['November']);
sns.kdeplot(wholedf['December']);
plt.title("Distribution of Departures by Month")
plt.xlim(0,100000)
plt.show()
```

Seasonal Distribution
--------------
Interestingly there is no distinct jump for oliday months and the distribution has a disctint two hump distribution.  Beaking it into sets of three and all set to the sameaxis limits you can see that November and December don't have a jump but July, August and October do.  

```python
sns.kdeplot(wholedf['January ']);
sns.kdeplot(wholedf['February']);
sns.kdeplot(wholedf['March']);
sns.kdeplot(wholedf['April']);
plt.title("Distribution of Departures by Month")
plt.xlim(0,100000)
plt.show()
```

```python
sns.kdeplot(wholedf['May']);
sns.kdeplot(wholedf['June']);
sns.kdeplot(wholedf['July']);
sns.kdeplot(wholedf['August']);
plt.title("Distribution of Departures by Month")
plt.xlim(0,100000)
plt.show()
```

```python
sns.kdeplot(wholedf['September']);
sns.kdeplot(wholedf['October']);
sns.kdeplot(wholedf['November']);
sns.kdeplot(wholedf['December']);
plt.title("Distribution of Departures by Month")
plt.xlim(0,100000)
plt.show()
```

Monthly Values over the Years
---------------
If we compare the actual plot of each month's value again you can see that July, August and October have consistantly being higher since the beginning.  Stacking the values in the bar chart allows you to see the same values in a different way.  The distribution seems to be fairly consistant over the years.

```python
plt.figure(); wholedf.plot();
plt.title("Value of Departures by Month")
plt.show()
```

```python
wholedf.plot.bar(stacked=True);
plt.title("Stacked Values of Departures by Year")
plt.ylim(0,1200000)
plt.show()
```

```python
yearsum = wholedf.sum(axis=1)

plt.plot(yearsum)
plt.title("Total Count of Departures by Year")
plt.ylim(0,800000)
plt.show()
```
